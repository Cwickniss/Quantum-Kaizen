```python
print("Quantum Solitaire is a single-player card game.\n")
print("In this prototype we use a simplified deck:")
print(" - Cards are numbered from 1 to 12 (no picture cards);")
print(" - There are two suits, called 0 and 1;\n")
print("In each round, you'll be given a hand of four face down cards, layed out in a square.\n")
print("You then get to choose two of them to keep, one on each side.\n")
print("The aim is to collect a card of each type.\n")
```

```python
print("The cards of each hand are random, except for a few rules.\n\n")
print("1. The two bottom cards are never both suit 1.\n")
print("2. The two top cards are never both odd numbers.\n")
print("3. If the top card on one side is of suit 1, the bottom card on the other is also of suit 1.\n")
print("4. If the top card on one side has an odd number, the bottom card on the other is also odd.\n")
print("Use these rules to maximize your chances of getting the cards you need.\n")
```

```python
# It's time for the main game loop.

# In this we deal hands, ask the players to choose cards, rinse and repeat

# the game variable will be set to False once the game is over
game = True

# the variable collection[X][Y] will tell us whether or not the player has card number Y of suit colour X (X=0 is black, X=1 is red)
collection = [ [False]*13 for _ in range(2)] # all values are initialized to false

# count number of rounds
round = 0

while (game):
    
    round = round + 1
    
    input("Press Enter to play hand number " + str(round) + "...\n")
    
    # generate dictionary to store cards for the hand
    hand = {'bl': { 'color': random.randint(0,1), 'number': random.randint(1,12) },
            'br': { 'color': random.randint(0,1), 'number': random.randint(1,12) },
            'tl': { 'color': random.randint(0,1), 'number': random.randint(1,12) },
            'tr': { 'color': random.randint(0,1), 'number': random.randint(1,12) }}
    
    # now we need to enforce the rules
    
    # rule 1
    if (hand['bl']['color']==1):
        hand['br']['color'] = 0
    # and the alt rule 1, which is a consequence of rules 1 and 3
    if (hand['tl']['color']==1):
        hand['tr']['color'] = 0
        
    # rule 2
    if ( (hand['tl']['number']%2)==1 ):
        hand['tr']['number'] = random.choice ( [2, 4, 6, 8, 10, 12] )
    # and the alt rule 2, which is a consequence of rules 2 and 4
    if ( (hand['bl']['number']%2)==1 ):
        hand['br']['number'] = random.choice ( [2, 4, 6, 8, 10, 12] )
    
    # rule 3 (and 1)
    if (hand['tl']['color']==1):
        hand['br']['color'] = 1
        hand['bl']['color'] = 0
    if (hand['tr']['color']==1):
        hand['bl']['color'] = 1
        hand['br']['color'] = 0
    
    # rule 4 (and 2)
    if ( (hand['bl']['number']%2)==1 ):
        hand['tr']['number'] = random.choice( [1, 3, 5, 7, 9, 11] )
        hand['tl']['number'] = random.choice( [2, 4, 6, 8, 10, 12] )
    if ( (hand['br']['number']%2)==1 ):
        hand['tl']['number'] = random.choice( [1, 3, 5, 7, 9, 11] )
        hand['tr']['number'] = random.choice( [2, 4, 6, 8, 10, 12] )
    
    print(str(hand['tl']) + "   " + str(hand['tr'])) # for debugging
    print(str(hand['bl']) + "   " + str(hand['br'])) # for debugging
    
    # ask user what two cards they want
    lastCard = ''
    for n in [0,1]:
        choosing = True
        while (choosing):
            
            if (round==1):
                card = input("Which card would you like to turn over?\nEnter tl for the top-left card, br for bottom-right, and so on.\n")
            else:
                card = input("Which card would you like to turn over? (tl, tr, bl or br)\n")
                
            if card in ['br','bl','tr','tl']:
                if ( ( card=='tr' or card=='br' ) and ( lastCard=='tr' or lastCard=='br' ) ) or ( ( card=='tl' or card=='bl' ) and ( lastCard=='tl' or lastCard=='bl' ) ):
                    print("\nYou took a card on that side. Try again.\n")
                else:
                    choosing = False
                    lastCard = card
            else :
                print("\nThat's not a valid card. Try again.\n")

        # display card
        print("\nHere's your card")
        print(hand[card])
        print("\n")

        # add to collection
        collection[ hand[card]['color'] ][ hand[card]['number'] ] = True
    
    input("Press Enter to see the cards you have so far.\n")
    for c in [0,1]:
        for n in range(1,13):
            if collection[c][n]:
                print ("color = " + str(c)+ ", number = " + str(n))
    print("\n")
    
    if False not in collection[0] and False not in collection[1]:
        print("You got all the cards in " + str(round) + " rounds.")
        print("Well done!")
        game  = False
        
```
